```python
import numpy as np
from scipy import signal
```

```python
data = np.loadtxt("minions_res.txt")
```

```python
data.shape
```

```python
grayscale=np.average(data,axis=1)
grayscale=grayscale.reshape(240,320)
```

```python
grayscale.shape
```

```python
Gx=np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
Gy=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
gradx = signal.convolve2d(grayscale, Gx,mode='same')
grady = signal.convolve2d(grayscale, Gy,mode='same')
```

```python
gradx.shape
```

```python
G=(gradx**2+grady**2)**(1/2)
G.shape
```

```python
print(np.amax(G))
print(np.amin(G))
```

```python
# binary
G[G<=127]=0
G[G>127]=1
```

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.imshow(G, cmap='binary', vmin=0, vmax=1)
plt.show()
```
